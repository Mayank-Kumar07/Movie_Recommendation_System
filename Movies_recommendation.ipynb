```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('imdb_top_1000.csv')
```

```python
data = data.drop(['Poster_Link','Certificate','Runtime','Meta_score','No_of_Votes'],axis=1)
```

```python
data.info
```

```python
data = data.rename(columns={'Series_Title':'Movie_Name'})
```

```python
data.head(5)
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
```

```python
def preprocess(text):
    if pd.isna(text):
        return []
    text = text.lower()
    tokens = word_tokenize(text)
    return tokens
```

```python
data['tokens'] = data.apply(lambda row: preprocess(f"{row['Movie_Name']} {row['Genre']} {row['Overview']} {row['Director']} {row['Star1']} {row['Star2']} {row['Star3']} {row['Star4']}"), axis = 1)
```

```python
data.tokens.head(10)
```

```python
sentences = data.tokens.tolist()
```

```python
sentences
```

```python
word2vec_model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)
```

```python
word2vec_model.save("Word2Vec_Movie_Model.model")
```

```python
def gen_embedding(tokens, model):
    vectors = [model.wv[token] for token in tokens if token in model.wv]
    if vectors:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(model.vector_size)
```

```python
data['embeddings'] = data.tokens.apply(lambda tokens: gen_embedding(tokens, word2vec_model))
```

```python
data.embeddings.head(10)
```

```python
def keyword_embedding(keywords, model):
    tokens = preprocess(keywords)
    vectors = [model.wv[token] for token in tokens if token in model.wv]
    if vectors:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(model.vector_size)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_movies_by_keywords(keywords, num_recommendations=5):
    embedding_keyword = keyword_embedding(keywords, word2vec_model)
    similarity_scores = cosine_similarity([embedding_keyword], data.embeddings.tolist())[0]
    
    similarity_scores = list(enumerate(similarity_scores))
    
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    
    movie_indices = [i[0] for i in similarity_scores[:num_recommendations]]
    #print(type(similarity_scores))
    return data.Movie_Name.iloc[movie_indices]
```

```python
print(recommend_movies_by_keywords("Christopher Nolan"))
```
